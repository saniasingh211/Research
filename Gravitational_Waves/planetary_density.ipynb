```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
R_planet_earth = 1.44 #earth radii from transit analysis
print(f"Kepler-10b radius: {R_planet_earth} Earth radii")
```

```python
# Constants
G = 6.674e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30     # kg
M_earth = 5.972e24   # kg

# Kepler-10 system
K = 3.3              # m/s - velocity amplitude
M_star = 0.895       # solar masses
P = 0.8375243        # days

# Convert to SI
M_star_kg = M_star * M_sun
P_seconds = P * 24 * 3600

# Calculate planet mass
M_planet = (K * (M_star_kg**(2/3)) * (P_seconds**(1/3))) / ((2 * np.pi * G)**(1/3))
M_planet_earth = M_planet / M_earth

print(f"Planet mass: {M_planet:.3e} kg")
print(f"Planet mass: {M_planet_earth:.2f} Earth masses")
```

```python
# Calculate density
R_earth_m = 6.371e6      # metres
M_earth_kg = 5.972e24    # kg

R_planet_m = 1.44 * R_earth_m    # our measurement
M_planet_kg = 3.33 * M_earth_kg  # published mass

volume = (4/3) * np.pi * R_planet_m**3
density = M_planet_kg / volume

# Earth's density for comparison
R_earth_m2 = 6.371e6
M_earth_kg2 = 5.972e24
volume_earth = (4/3) * np.pi * R_earth_m2**3
density_earth = M_earth_kg2 / volume_earth

print(f"Kepler-10b density: {density:.0f} kg/m³")
print(f"Earth density: {density_earth:.0f} kg/m³")
print(f"Ratio: {density/density_earth:.2f} x Earth")

# Reference densities
print(f"\nFor reference:")
print(f"Water: 1000 kg/m³")
print(f"Rock: ~3000 kg/m³")
print(f"Iron: ~7800 kg/m³")
```
