```python
import numpy as np

# What we measured
R_planet_earth = 1.44   # Earth radii (from transit)

M_planet_earth = 4.52   # Earth masses (from radial velocity)

# Earth reference values in SI
R_earth = 6.371e6    # metres
M_earth = 5.972e24   # kg

# Convert to SI
R = R_planet_earth * R_earth   # metres
M = M_planet_earth * M_earth   # kg

print(f"Radius: {R:.3e} m")
print(f"Mass:   {M:.3e} kg")
```

```python
# Calculate volume
V = (4/3) * np.pi * R**3

# Calculate density
rho = M / V

# Convert to g/cm³ (standard units for planetary density)
rho_gcc = rho / 1000  # kg/m³ → g/cm³

print(f"Volume:  {V:.3e} m³")
print(f"Density: {rho:.1f} kg/m³")
print(f"Density: {rho_gcc:.2f} g/cm³")
print(f"Published value: 8.80 g/cm³")
```

```python
# Context
earth_density = 5.51  # g/cm³
water_density = 1.00
iron_density = 7.87

print(f"Kepler-10b density: {rho_gcc:.2f} g/cm³")
print(f"Earth density:      {earth_density} g/cm³")
print(f"Iron density:       {iron_density} g/cm³")
print(f"Water density:      {water_density} g/cm³")
print("Result:")
print(f"Kepler-10b is {rho_gcc/earth_density:.2f}x denser than Earth")
print(f"Kepler-10b is {rho_gcc/iron_density:.2f}x denser than iron")
```
