```python
import numpy as np
import matplotlib.pyplot as plt
import radis
```

```python
from radis import calc_spectrum

s = calc_spectrum(
    wavenum_min=2200,    # much narrower — just the CO2 4.3 micron feature
    wavenum_max=2400,
    molecule='CO2',
    isotope='1',
    pressure=0.01,
    Tgas=1200,
    mole_fraction=0.01,
    databank='hitran',
    verbose=False
)

s.plot('absorbance')
```

```python
from radis import calc_spectrum

s = calc_spectrum(
    wavenum_min=2200,
    wavenum_max=2400,
    molecule='CO2',
    isotope='1',
    pressure=0.01,
    Tgas=1200,
    mole_fraction=0.01,
    databank='hitran',
    verbose=0,
    warnings={'AccuracyError': 'ignore',
              'AccuracyWarning': 'ignore'}
)

s.plot('absorbance')
plt.show()
```
