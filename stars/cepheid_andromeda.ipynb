```python
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hubble's original Cepheid in Andromeda — M31-V1
period_m31 = 31.91          # days
apparent_mag = 18.5         # V band apparent magnitude

# your Leavitt Law from LMC
slope = -2.917
intercept = 16.844

# predicted true magnitude from Leavitt Law
true_mag = slope * np.log10(period_m31) + intercept

# distance modulus
distance_modulus = apparent_mag - true_mag

# convert to light years
# distance modulus μ = 5*log10(d/10pc)
distance_pc = 10 ** ((distance_modulus / 5) + 5)
distance_ly = distance_pc * 3.262

print(f"Period: {period_m31} days")
print(f"True magnitude (Leavitt Law): {true_mag:.2f}")
print(f"Apparent magnitude (telescope): {apparent_mag}")
print(f"Distance modulus: {distance_modulus:.2f}")
print(f"Distance: {distance_pc/1000:.0f} kpc")
print(f"Distance: {distance_ly/1e6:.2f} million light years")
```
